# Converting Data Into A List Of Lists

```python
def read_csv(filename):
    data_string = open(filename).read()
    data_list = data_string.split("\n")[1:]
    final_list = list()
    
    for row in data_list:
        row_list = row.split(",")
        int_row_list = list()
        for val_string in row_list:
            int_row_list.append(int(val_string))
        final_list.append(int_row_list)
    return final_list
        
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
```

```python
cdc_list[0:10]
```

# Calculating Number Of Births Each Month

```python
def month_births(data):
    month_sums = dict()
    
    for row in data:
        month = row[1]
        births = row[4]
        if month in month_sums:
            month_sums[month] = month_sums[month] + births
        else:
            month_sums[month] = births
    return month_sums
    
cdc_month_births = month_births(cdc_list)
```

```python
cdc_month_births
```

# Calculating Number Of Births Each Day Of Week

```python
def dow_births(data):
    dow_sums = dict()
    
    for row in data:
        dow = row[3]
        births = row[4]
        if dow in dow_sums:
            dow_sums[dow] = dow_sums[dow] + births
        else:
            dow_sums[dow] = births
    return dow_sums
    
cdc_dow_births = dow_births(cdc_list)
```

```python
cdc_dow_births
```

# Creating A More General Function

```python
def calc_counts(data, col_num):
    sums_dict = dict()
    
    for row in data:
        col_value = row[col_num]
        births = row[4]
        if col_value in sums_dict:
            sums_dict[col_value] = sums_dict[col_value] + births
        else:
            sums_dict[col_value] = births
    return sums_dict

cdc_year_births = calc_counts(cdc_list, 0)
cdc_month_births = calc_counts(cdc_list, 1)
cdc_dom_births = calc_counts(cdc_list, 2)
cdc_dow_births = calc_counts(cdc_list, 3)
```
